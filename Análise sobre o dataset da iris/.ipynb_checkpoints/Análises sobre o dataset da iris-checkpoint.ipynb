## Importando os dados

```python
# Importando o dataset
from sklearn import datasets

# Carregando os dados na variável iris e definindo-os como dataframe
iris = datasets.load_iris(as_frame=True)
```

```python
# Salvando os dados na variável X
X = iris.data
```

Seria interessante adicionar ao dataframe **duas colunas** contendo o **nome das espécies** e o **número que as identifica**. Para isso, vamos criar uma lista contendo o nome de cada espécie de acordo com as features de cada linha.

```python
species_list = []                              # Lista para salvar o nome das espécies
for i in iris.target:                          # Iterando sobre o número que identifica cada espécie 
    species_list.append(iris.target_names[i])  # Adicionando a espécia à lista de acordo com o número da espécie
```

Criando as duas novas colunas

```python
X['Species'] = species_list
```

```python
X['Species_number'] = iris.target
```

```python
# Mostrando o estado atual do dataframe
X
```

## Visualizando as informações

Gráfico mostrando cada categoria em função das variáveis **sepal length** (comprimento da sépala) e **petal length** (comprimento da pétala).

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4,4))

graph1 = ax.scatter(x=X['sepal length (cm)'], y=X['petal length (cm)'], c=X['Species_number']);

handles, labels = graph1.legend_elements();
ax.legend(handles, iris.target_names, loc="upper left", title="Classes");

plt.xlabel('sepal length (cm)');
plt.ylabel('petal length (cm)');
```

Mostrando o **intervalo de valores** que podem ser assumidos **por cada categoria** em função de **sepal length, sepal width, petal length e petal width**.

```python
import seaborn as sns

fig, axs = plt.subplots(ncols=4, figsize=(18,6))
sns.stripplot(x='Species',y='sepal length (cm)',data=X.drop('Species_number', axis=1),jitter=True,edgecolor='gray',ax=axs[0]);
sns.stripplot(x='Species',y='sepal width (cm)',data=X.drop('Species_number', axis=1),jitter=True,edgecolor='gray', ax=axs[1]);
sns.stripplot(x='Species',y='petal length (cm)',data=X.drop('Species_number', axis=1),jitter=True,edgecolor='gray', ax=axs[2]);
sns.stripplot(x='Species',y='petal width (cm)',data=X.drop('Species_number', axis=1),jitter=True,edgecolor='gray', ax=axs[3]);
```

Mostrando a mesma informação através do **gráfico de violino**.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20,15))

sns.violinplot(y='Species', x='sepal length (cm)', data=X, ax=axs[0][0])
sns.violinplot(y='Species', x='sepal width (cm)', data=X, ax=axs[0][1])
sns.violinplot(y='Species', x='petal length (cm)', data=X, ax=axs[1][0])
sns.violinplot(y='Species', x='petal width (cm)', data=X, ax=axs[1][1])

plt.show()
```

## Separando em treino e teste

```python
# Separando em X_data as variáveis que farão a predição
X_data = X.drop(['Species','Species_number'], axis=1)
X_data.head(3)
```

```python
# Separando em y_data a coluna que contem as labels numéricas (o valor que representa cada categoria)
y_data = X['Species_number']
y_data
```

```python
from sklearn.model_selection import train_test_split

# Separando 30% dos dados para teste e 70% para treino
X_train, X_test, y_train, y_test = train_test_split(X_data,y_data,test_size = 0.3)
```

```python
# Mostrando a quantidade de dados reservados para treino e teste, respectivamente
(X_train.shape[0], X_test.shape[0])
```

```python
# Somando as duas quantidades obtêm-se as 150 linhas do dataframe original
X_train.shape[0] + X_test.shape[0]
```

## Treinando modelos

**Regressão Logística**

Treinando o modelo de regressão logística.

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import warnings

warnings.filterwarnings('ignore')

log_ref = LogisticRegression(random_state=0)
log_ref.fit(X_train,y_train)
```

Fazendo as predições.

```python
y_pred = log_ref.predict(X_test)
```

**Acurácia** obtida (porcentagem que representa o quanto o modelo acertou)

```python
metrics.accuracy_score(y_test, y_pred)
```

**Visualizando graficamente** as **instâncias** (representadas por pontos no gráfico) **preditas erroneamente** pelo modelo.

```python
fig, axs = plt.subplots(ncols=2, figsize=(10,5))
figura1 = axs[0].scatter(x=X_test['sepal length (cm)'], y=X_test['petal length (cm)'], c=y_pred);
figura2 = axs[1].scatter(x=X_test['sepal length (cm)'], y=X_test['petal length (cm)'], c=y_test);

axs[0].set_xlabel('sepal length (cm)');
axs[0].set_ylabel('petal length (cm)');
axs[1].set_xlabel('sepal length (cm)');
axs[1].set_ylabel('petal length (cm)');
axs[0].set_title('Previsões feitas nos dados de teste');
axs[1].set_title('Dados de teste verdadeiros');

handles, labels = figura1.legend_elements();
axs[0].legend(handles, iris.target_names, loc="lower right", title="Classes");
axs[1].legend(handles, iris.target_names, loc="lower right", title="Classes");
```

**KNN (K-Nearest Neighbors)**

Treinando o modelo **KNN**. Nesse caso, **testaremos 26 parâmetros diferentes** para **n_neighbors** (de **1 a 26**) e salvaremos os resultados na lista scores.

```python
import warnings
from sklearn.neighbors import KNeighborsClassifier

warnings.filterwarnings('ignore')

scores = []
for k in list(range(1,26)):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
import numpy as np

# Os scores obtidos para cada parâmetro de n_neighbors (1 a 26)
np.array(scores)
```

```python
# Maior pontuação obtida
np.array(scores).max()
```

```python
# Primeira posição em que se encontra no vetor
posicao = next(i for i, x in enumerate(np.array(scores)) if x == np.array(scores).max())
```

```python
# Vendo as predições
knn = KNeighborsClassifier(n_neighbors=(posicao+1))
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
y_pred
```

**Visualizando graficamente** as **instâncias** (representadas por pontos no gráfico) **preditas erroneamente** pelo modelo.

```python
fig, axs = plt.subplots(ncols=2, figsize=(10,4))

axs[0].scatter(x=X_test['sepal length (cm)'], y=X_test['petal length (cm)'], c=y_pred);
axs[1].scatter(x=X_test['sepal length (cm)'], y=X_test['petal length (cm)'], c=y_test);
axs[0].set_xlabel('sepal length (cm)');
axs[0].set_ylabel('petal length (cm)');
axs[1].set_xlabel('sepal length (cm)');
axs[1].set_ylabel('petal length (cm)');
axs[0].set_title('Previsões feitas nos dados de teste');
axs[1].set_title('Dados de teste verdadeiros');

handles, labels = figura1.legend_elements();
axs[0].legend(handles, iris.target_names, loc="lower right", title="Classes");
axs[1].legend(handles, iris.target_names, loc="lower right", title="Classes");
```
