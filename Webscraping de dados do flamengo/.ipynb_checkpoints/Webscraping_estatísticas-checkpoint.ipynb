## Objetivo

Pegar informações do nome dos jogadores do flamengo e da quantidade de gols fazendo uma requisição no site: 'https://www.espn.com.br/futebol/time/estatisticas/_/id/819/flamengo'.
Após isso, plotar um gráfico com os dados. A vantagem de fazer isso é que a tabela e o gráfico gerados são atualizados automaticamente quando o site atualiza as informações!

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

html_doc = requests.get('https://www.espn.com.br/futebol/time/estatisticas/_/id/819/flamengo')
soup = BeautifulSoup(html_doc.text, 'html.parser')
#print(html_doc.text)
print(soup.prettify())
```

```python
jogadores_lista = []
for jogador in soup.tbody.find_all('tr'):
    #print(jogador.a.text)
    jogadores_lista.append(jogador.a.text)
```

```python
print(jogadores_lista)
```

```python
gols_lista = []

for i in range(2,len(soup.tbody.find_all('span')),3):
    #print(soup.tbody.find_all('span')[i].text)
    gols_lista.append(soup.tbody.find_all('span')[i].text)
    
print(gols_lista)
```

```python
dados_jogadores = []
for i in range(len(jogadores_lista)):
    dados_jogadores.append([jogadores_lista[i],gols_lista[i]])
```

```python
print(dados_jogadores)
```

```python
tabela = pd.DataFrame(dados_jogadores, columns = ["Nome do jogador", "Número de gols"])
```

```python
display(tabela)
```

```python
for i in range(tabela.shape[0]):
    if int(tabela.loc[i, 'Número de gols']) == 0:
        tabela = tabela.drop(i,axis=0)
```

```python
import plotly.express as px

fig = px.pie(tabela, values="Número de gols", names="Nome do jogador", title="Gols do Flamengo no Brasileirão", width=500, height=500)
fig.show()
```
